```python
def convert_length(value, from_unit):
    """
    Converts length between meters and feet.
    :param value: Numeric value to convert
    :param from_unit: 'm' for meters to feet, 'ft' for feet to meters
    :return: Converted value
    """
    conversion_factor = 3.28084
    if from_unit == 'm':
        return value * conversion_factor
    elif from_unit == 'ft':
        return value / conversion_factor
    else:
        raise ValueError("Unsupported length unit")

def convert_weight(value, from_unit):
    """
    Converts weight between kilograms and pounds.
    :param value: Numeric value to convert
    :param from_unit: 'kg' for kilograms to pounds, 'lbs' for pounds to kilograms
    :return: Converted value
    """
    conversion_factor = 2.20462
    if from_unit == 'kg':
        return value * conversion_factor
    elif from_unit == 'lbs':
        return value / conversion_factor
    else:
        raise ValueError("Unsupported weight unit")

def convert_volume(value, from_unit):
    """
    Converts volume between liters and gallons.
    :param value: Numeric value to convert
    :param from_unit: 'L' for liters to gallons, 'gal' for gallons to liters
    :return: Converted value
    """
    conversion_factor = 0.264172
    if from_unit == 'L':
        return value * conversion_factor
    elif from_unit == 'gal':
        return value / conversion_factor
    else:
        raise ValueError("Unsupported volume unit")

def main():
    try:
        print("Choose conversion type:")
        print("1. Length (m <-> ft)")
        print("2. Weight (kg <-> lbs)")
        print("3. Volume (L <-> gal)")

        choice = input("Enter choice (1/2/3): ")
        value = float(input("Enter value to convert: "))

        if choice == '1':
            unit = input("Enter unit ('m' for meters, 'ft' for feet): ")
            result = convert_length(value, unit)
            print("Converted value:", result)
        elif choice == '2':
            unit = input("Enter unit ('kg' for kilograms, 'lbs' for pounds): ")
            result = convert_weight(value, unit)
            print("Converted value:", result)
        elif choice == '3':
            unit = input("Enter unit ('L' for liters, 'gal' for gallons): ")
            result = convert_volume(value, unit)
            print("Converted value:", result)
        else:
            print("Invalid choice.")
    except ValueError as e:
        print("Error:", e)
    except Exception as e:
        print("Unexpected error:", e)

if __name__ == "__main__":
    main()
```

```python
def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.
    :param numbers: List of numeric values
    :return: Sum of numbers
    """
    return sum(numbers)

def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.
    :param numbers: List of numeric values
    :return: Average of numbers
    """
    return sum(numbers) / len(numbers) if numbers else 0

def find_maximum(numbers):
    """
    Finds the maximum number in a list.
    :param numbers: List of numeric values
    :return: Maximum value
    """
    return max(numbers)

def find_minimum(numbers):
    """
    Finds the minimum number in a list.
    :param numbers: List of numeric values
    :return: Minimum value
    """
    return min(numbers)

def main():
    try:
        print("Choose an operation:")
        print("1. Sum")
        print("2. Average")
        print("3. Maximum")
        print("4. Minimum")

        choice = input("Enter choice (1/2/3/4): ")
        numbers = list(map(float, input("Enter numbers separated by spaces: ").split()))

        if not numbers:
            raise ValueError("The list of numbers cannot be empty.")

        if choice == '1':
            print("Sum:", calculate_sum(numbers))
        elif choice == '2':
            print("Average:", calculate_average(numbers))
        elif choice == '3':
            print("Maximum:", find_maximum(numbers))
        elif choice == '4':
            print("Minimum:", find_minimum(numbers))
        else:
            print("Invalid choice.")
    except ValueError as e:
        print("Error:", e)
    except Exception as e:
        print("Unexpected error:", e)

if __name__ == "__main__":
    main()

```

```python
def extract_every_other(lst):
    """
    Extracts every other element from the list, starting from the first element.
    :param lst: List of elements
    :return: List with every other element
    """
    return lst[::2]

def get_sublist(lst, start, end):
    """
    Returns a sublist from the start index to the end index (inclusive).
    :param lst: List of elements
    :param start: Starting index
    :param end: Ending index (inclusive)
    :return: Sublist from start to end index
    """
    return lst[start:end+1]

def reverse_list(lst):
    """
    Reverses the list using slicing.
    :param lst: List of elements
    :return: Reversed list
    """
    return lst[::-1]

def remove_first_last(lst):
    """
    Removes the first and last elements of the list.
    :param lst: List of elements
    :return: List without first and last elements
    """
    return lst[1:-1]

def get_first_n(lst, n):
    """
    Extracts the first n elements from the list.
    :param lst: List of elements
    :param n: Number of elements to extract
    :return: List with first n elements
    """
    return lst[:n]

def get_last_n(lst, n):
    """
    Extracts the last n elements from the list.
    :param lst: List of elements
    :param n: Number of elements to extract
    :return: List with last n elements
    """
    return lst[-n:]

def reverse_skip(lst):
    """
    Extracts elements in reverse order, starting from the second-to-last and skipping one in between.
    :param lst: List of elements
    :return: List of elements in reverse order with a skip
    """
    return lst[-2::-2]

def main():
    sample_list = [1, 2, 3, 4, 5, 6]
    print("Every other element:", extract_every_other(sample_list))
    print("Sublist (2 to 4):", get_sublist(sample_list, 2, 4))
    print("Reversed list:", reverse_list(sample_list))
    print("Without first and last:", remove_first_last(sample_list))
    print("First 3 elements:", get_first_n(sample_list, 3))
    print("Last 2 elements:", get_last_n(sample_list, 2))
    print("Reverse skip:", reverse_skip(sample_list))

if __name__ == "__main__":
    main()
```

```python
import numpy as np

# 1. Prove A * A_inv = I
A = np.array([[3, 4], [7, 8]])
A_inv = np.linalg.inv(A)

# Identity matrix
I = np.dot(A, A_inv)

print("1. Prove A * A_inv = I (Identity matrix):")
print("A * A_inv = \n", I)

# 2. Prove AB != BA
B = np.array([[5, 3], [2, 1]])

# Compute AB and BA
AB = np.dot(A, B)
BA = np.dot(B, A)

print("\n2. Prove AB != BA:")
print("AB = \n", AB)
print("BA = \n", BA)

# Check if AB equals BA
print("Is AB equal to BA?", np.array_equal(AB, BA))

# 3. Prove (AB)^T = B^T @ A^T
AB_T = np.transpose(AB)
B_T_A_T = np.dot(np.transpose(B), np.transpose(A))

print("\n3. Prove (AB)^T = B^T @ A^T:")
print("Transpose of AB = \n", AB_T)
print("B.T @ A.T = \n", B_T_A_T)

# Check if (AB)^T equals B.T @ A.T
print("Is (AB)^T equal to B.T @ A.T?", np.array_equal(AB_T, B_T_A_T))

# 4. Solve the system of linear equations using Inverse Method
# The system of equations is: 2x - 3y + z = -1, x - y + 2z = -3, 3x + y - z = 9
# Represent this system in matrix form: AX = B
A_system = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_system = np.array([[-1], [-3], [9]])

# Solve for X using the inverse method: X = A_inv * B
A_inv_system = np.linalg.inv(A_system)
X_system = np.dot(A_inv_system, B_system)

print("\n4. Solve the system using Inverse Method:")
print("Solution for x, y, z: \n", X_system)

# 5. Solve the system using np.linalg.solve function
X_direct = np.linalg.solve(A_system, B_system)

print("\n5. Solve the system using np.linalg.solve:")
print("Direct solution for x, y, z: \n", X_direct)
```
